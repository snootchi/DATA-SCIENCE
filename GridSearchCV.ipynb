##**WITHOUT PIPELINE**

```python
import pandas as pd
import numpy as np 
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import GridSearchCV,train_test_split,RandomizedSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('/content/Telecom Churn Data SingTel.csv')
df.head()
```

```python
df.drop('Phone Number',axis=1 , inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
#Transforming the Categorical Columns into Numerical columns
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Churn']=label_encoder.fit_transform(df['Churn'])
df['State'] = label_encoder.fit_transform(df['State'])
df['International Plan'] = label_encoder.fit_transform(df['International Plan'])
df['Voice mail Plan'] = label_encoder.fit_transform(df['Voice mail Plan'])
df.head()
```

```python
x = df.drop('Churn' , 1)
y = df['Churn']
y.head()
```

```python
y.value_counts(normalize=True)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2 , random_state=0 , stratify=y)
```

```python
df.info
```

```python
cat_cols = ['State' , 'International Plan' , 'Voice mail Plan'] #categorical columns
#numerical columns
num_cols = ['Account Length','Area Code', 'Num of Voice mail Messages','Total Day Minutes', 'Total Day Calls', 'Total day Charge','Total Eve Minutes', 'Total Eve Calls', 'Total Eve Charge', 'Total Night Minutes', 'Total Night Calls ', 'Total Night Charge', 'Total International Minutes', 'Total Intl  Calls', 'Total Intl Charge','Number Customer Service calls ']
```

```python
df.columns
```

Rescaling the Numerical Columns to Standarization >>> N(0,1) 

```python
ss = StandardScaler()
x_train_num = ss.fit_transform(x_train[num_cols])
x_train_num[:5]
```

```python
x_test_num = ss.fit_transform(x_test[num_cols])
x_test_num[:5]
```

We'll concatenate/Add the X_train_num with X_train of the categorical columns to finally give the Final x_train_out output dataframe

```python
x_train_out = pd.DataFrame(np.concatenate([x_train_num , x_train[cat_cols]],axis=1))
x_train_out.head()
```

```python
x_test_out = pd.DataFrame(np.concatenate([x_test_num,x_test[num_cols]],axis=1))
x_test_out.head()
```

We've Now the test and the training data Ready !! \
Moving to the **Random Forest Classifier** by initiating the **rfc** and put the paramaters multiple choice to find the optimal one

```python
rfc = RandomForestClassifier(random_state=0)
params = {
    'n_estimators' : [10,20],
    'max_depth' : [2,3]
}
```

Moving to the **GridSearchCV** which we're going to use it 

```python
gs = GridSearchCV(estimator=rfc , param_grid=params , cv=5 , n_jobs=-1)
```

**n_jobs** : mean how many compututions we want , is the Number of jobs to run in parallel , **-1** means we using all processus so all the **combinations** between parameters .

```python
gs.fit(x_train_out , y_train)
```

let's Look at the **Result** of GridSearchCV

```python
pd.DataFrame(gs.cv_results_)
```

we gonna see the several parameters which the model been evaluating , and in the end it'll giving us the **rank_tes_score** according to the model \
Or we can do simply **gs.best_params_** it will give us the optimal parameters for the model

```python
gs.best_params_
```

we gonna utilize the previous optimal parameters that we obtain and fitting over the x and y training .

```python
rf = RandomForestClassifier(random_state=0 , **gs.best_params_)
```

```python
rf.fit(x_test_out , y_test)
```

```python
rf.score(x_test_out , y_test)
```

the score the we obtain is **91.2%** occurancy 

##**WITH PIPELINE**

```python
x_train.columns
```

```python
ct = ColumnTransformer([('ss' , StandardScaler() , num_cols)] , remainder='passthrough')
```

we'll fit our X train

```python
ct.fit(x_train)
```

```python
ct.transformers_
```

it passes through all the categorical cols which we've mentioned , \
we'll Creating the Pipeline with our training data and then applying RandomForestClassifier

```python
from scipy.sparse.construct import rand
pipe = Pipeline([
                 ('ct' , ct),
                 ('model' , RandomForestClassifier(random_state=0 , n_jobs=-1))
])
```

```python
params_new = {
    'model__n_estimators' : [10,20],
    'model__max_depth': [2,3],
    'ct__ss__with_mean' : [True,False]
}
```

```python
gs_new = GridSearchCV(estimator=pipe , param_grid=params_new , n_jobs=-1)
```

```python
x_train.head()
```

then fitting our data on new gs

```python
x_train.head()
```

```python
gs_new.fit(x_train,y_train)
```

```python
gs_new.best_params_
```

```python
gs_new.best_score_
```
