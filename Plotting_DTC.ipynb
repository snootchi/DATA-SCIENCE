```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

we're importing the 'iris' data set from sklearn and importing **the tree** for using DecisionTreeClassifier

```python
from sklearn.datasets import load_iris 
from sklearn import tree 
```

loading the iris data set with iris , and fit the iris data with iris target where : \
iris['data'] : is the X-Axis which contains all features \
iris['target'] : is the Y-Axis which contains classes 

```python
clf = tree.DecisionTreeClassifier(random_state=0)
iris = load_iris()

clf = clf.fit(iris['data'] , iris['target'])
```

```python
iris['target']
```

iris['target'] contains 3 **classes** {0,1,2} and we performed Tree Decision Classifier over it 

```python
plt.figure(figsize=(18,10))
tree.plot_tree(clf, filled=True);
```

we can also just use the code below to have the textual information about the same graph

```python
print(tree.export_text(clf))
```
