```python
import plotly.express as px
fig = px.scatter(df, x="area", y="price", color="basement")
```

```python
fig.show()
```

```python
fig = px.scatter(df, x="area", y="price", color="bedrooms" , marginal_x="histogram", marginal_y="histogram")
fig.show()
```

```python
#With a linear regression
fig = px.scatter(df, x="area", y="price", trendline="ols")
fig.show()
```

## **Pie Chartes** 

```python
fig = px.pie(df, values='price', names='bedrooms',title='Percentage of Prices based on bedrooms')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
services = ['Compt&Fin' , 'GRH','Marketing','Transportation']
amount = [450,300,500,450]
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']

trace = go.Pie(labels=services , values=amount,hoverinfo='percent+label',textinfo='value',textfont=dict(size=25),
              marker = dict(colors=colors , line=dict(color='#000000',width=1.5)))
#plot
iplot([trace])
```
